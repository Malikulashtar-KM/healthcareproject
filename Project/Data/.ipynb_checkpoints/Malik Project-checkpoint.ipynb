```python
# liteate male
```

```python
literatemale=census1q5[census1q5['Literate_Male'].isna()].index
census1q5.loc[literatemale,'Literate_Male'] = census1q5.loc[literatemale,'Literate'] - census1q5.loc[literatemale,'Literate_Female']
```

```python
#literate Female
```

```python
literatefemale=census1q5[census1q5['Literate_Female'].isna()].index
census1q5.loc[literatefemale,'Literate_Female'] = census1q5.loc[literatefemale,'Literate'] - census1q5.loc[literatefemale,'Literate_Male']
```

```python
# Households_Rural 
```

```python
Households_Rural=census1q5[census1q5['Households_Rural'].isna()].index
census1q5.loc[Households_Rural,'Households_Rural'] = census1q5.loc[Households_Rural,'Households'] - census1q5.loc[Households_Rural,'Urban_Households']
```

```python
census1q5.info()
```

```python
Householdsurban=census1q5[census1q5['Urban_Households'].isna()].index
census1q5.loc[Householdsurban,'Urban_Households'] = census1q5.loc[Householdsurban,'Households'] - census1q5.loc[Householdsurban,'Households_Rural']
```

```python
#HouseHold values
```

```python
Household=census1q5[census1q5['Households'].isna()].index
census1q5.loc[Household,'Households'] = census1q5.loc[Household,'Urban_Households'] + census1q5.loc[Household,'Households_Rural']
```

```python
#Population  = Young_and_Adult+  Middle_Aged + Senior_Citizen + Age_Not_Stated
```

```python
census1q5['Young_and_Adult'].fillna(census1q5['Population'] - (census1q5['Middle_Aged'] + census1q5['Senior_Citizen'] + census1q5['Age_Not_Stated']), inplace=True)
```

```python
census1q5['Middle_Aged'].fillna(census1q5['Population'] - (census1q5['Young_and_Adult'] + census1q5['Senior_Citizen'] + census1q5['Age_Not_Stated']), inplace=True)
```

```python
census1q5['Senior_Citizen'].fillna(census1q5['Population'] - (census1q5['Young_and_Adult'] + census1q5['Middle_Aged'] + census1q5['Age_Not_Stated']), inplace=True)
```

```python
census1q5['Age_Not_Stated'].fillna(census1q5['Population'] - (census1q5['Young_and_Adult'] + census1q5['Senior_Citizen'] + census1q5['Middle_Aged']), inplace=True)
```

## Question 7

```python
census1=census1q5.copy()
```

```python
dfq6=census1[['District','Households_Rural','Urban_Households','Households']]
dfq6.rename(columns={'Urban_Households':'Households_Urban'},inplace=True)
```

```python
dfq7 = pd.read_csv("housing.csv", usecols=['District Name','Rural/Urban','Total Number of households','Total Number of Livable','Total Number of Dilapidated','Latrine_premise'])
```

```python
dfq7.head(5)
```

```python
dfq7_rural = dfq7[dfq7['Rural/Urban']=='Rural']
dfq7_urban = dfq7[dfq7['Rural/Urban']=='Urban']
```

```python
dfq7_rural_final = dfq6.merge(dfq7_rural,left_on= 'District',right_on = 'District Name')
dfq7_rural_final['Total Number of Livable'] = dfq7_rural_final['Total Number of Livable']*dfq7_rural_final['Households_Rural']/100
dfq7_rural_final['Total Number of Dilapidated'] = dfq7_rural_final['Total Number of Dilapidated']*dfq7_rural_final['Households_Rural']/100
dfq7_rural_final['Latrine_premise'] = dfq7_rural_final['Latrine_premise']*dfq7_rural_final['Households_Rural']/100

dfq7_rural_final.drop(['Households_Urban','Rural/Urban','Households','District Name','Total Number of households'],axis = 1,inplace = True)

dfq7_rural_final.rename(columns={'Total Number of Livable':'Households_Rural_Livable','Total Number of Dilapidated':'Households_Rural_Dilapidated','Latrine_premise':'Households_Rural_Toilet_Premise'},inplace = True)
dfq7_rural_final

dfq7_urban_final = dfq6.merge(dfq7_urban,left_on= 'District',right_on = 'District Name')
dfq7_urban_final['Total Number of Livable'] = dfq7_urban_final['Total Number of Livable']*dfq7_urban_final['Households_Urban']/100
dfq7_urban_final['Total Number of Dilapidated'] = dfq7_urban_final['Total Number of Dilapidated']*dfq7_urban_final['Households_Rural']/100
dfq7_urban_final['Latrine_premise'] = dfq7_urban_final['Latrine_premise']*dfq7_urban_final['Households_Urban']/100

dfq7_urban_final.drop(['District','Households_Rural','Rural/Urban','Households','District Name','Total Number of households'],axis = 1,inplace = True)

dfq7_urban_final.rename(columns={'Total Number of Livable':'Households_Urban_Livable','Total Number of Dilapidated':'Households_Rrban_Dilapidated','Latrine_premise':'Households_Urban_Toilet_Premise'},inplace = True)
dfq7_urban_final
```

```python
dfq7_final_final = pd.concat([dfq7_rural_final, dfq7_urban_final], axis=1)
```

```python
dfq7_final_final.dropna(inplace=True)
```
